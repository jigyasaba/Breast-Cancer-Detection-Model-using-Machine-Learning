```python
import pandas as pd
import numpy as np
import os
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
from sklearn.preprocessing import LabelEncoder
from sklearn import tree 
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn import metrics
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import BaggingClassifier
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('breast-cancer.csv')
print(df.head())
```

```python
df.info()
```

```python
df.columns.values
print(df.columns)#printing the columns in dataset
#diagnosis is the columns we want to predict  as benign or malignant
```

```python
df.describe()
```

```python
print(df.shape)
```

```python
df.diagnosis
```

```python
print(df.diagnosis.value_counts())
```

```python
#missing values in dataset(Data Cleaning)
df.isnull().sum()
df.isna().sum()
```

```python
#Feature Encoding
# df['diagnosis'].replace(['B','M'],[0,1],inplace=True)
le=LabelEncoder()
#Feature selection
df['diagnosis']=le.fit_transform(df['diagnosis'])
X = df.drop(columns=['diagnosis'])#independent variables
y = df['diagnosis']#target variable
```

```python
print(X.shape)
print(y.shape)
```

```python
#splitting the datasetY->predictor
#X->input features 
train_X,test_X,train_y,test_y= train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
print(train_X.shape)
print(test_X.shape)
```

```python
#Decision Tree Classifier
model1=tree.DecisionTreeClassifier()
model1.fit(train_X,train_y)
```

```python
predict=model1.predict(test_X)
print("Accuracy:",accuracy_score(test_y,predict))
```

```python
print(model1.score(test_X,test_y))
print("Confusion Matrix:\n", confusion_matrix(test_y, predict))
```

```python
print('Accuracy score of the Decision Tree model is {:.2f}%'.format(accuracy_score(predict,test_y)*100))
print('F1 score of the Decision Tree model is {}'.format(f1_score(predict,test_y)))
```

```python
#Logistic Regression
model2=LogisticRegression()
model2.fit(train_X,train_y)
```

```python
print(model2.score(test_X,test_y))
predict=model2.predict(test_X)
print("Confusion Matrix:\n", confusion_matrix(test_y, predict))
```

```python

print("Accuracy:",accuracy_score(test_y,predict))
print('Accuracy score of the Logistic Regression model  is {:.2f}%'.format(accuracy_score(predict,test_y)*100))
print('F1 score of the Logistic Regression model is {}'.format(f1_score(predict,test_y)))
```

```python
model3=SVC()
model3.fit(train_X,train_y)
```

```python
print(model3.score(test_X,test_y))
predict=model3.predict(test_X)
print("Confusion Matrix:\n", confusion_matrix(test_y, predict))
```

```python
print("Accuracy:",accuracy_score(test_y,predict))
print('Accuracy score of the SVM model is {:.2f}%'.format(accuracy_score(predict,test_y)*100))
print('F1 score of the  SVM model is {}'.format(f1_score(predict,test_y)))
```

```python
model4=RandomForestClassifier()
model4.fit(train_X,train_y)
```

```python
print(model4.score(test_X,test_y))
predict=model4.predict(test_X)
print("Confusion Matrix:\n", confusion_matrix(test_y,predict))
print("Accuracy:",accuracy_score(test_y,predict))

```

```python
print("Accuracy score of the Random Forest Model is {:.2f}%".format(accuracy_score(test_y,predict)*100))
print("F1 score of the  SVM model is {}".format(f1_score(predict,test_y)))
```

```python
bagging_model = BaggingClassifier(estimator=tree.DecisionTreeClassifier(), n_estimators=10, random_state=0)
bagging_model.fit(train_X, train_y)
y_pred = bagging_model.predict(test_X)
```
